```python
from selenium import webdriver
import pandas as pd
from IPython.core.display import HTML
```

```python
fawzi=webdriver.Chrome()
fawzi.get("https://www.noon.com/egypt-en/electronics-and-mobiles/mobiles-and-accessories/mobiles-20905/apple/?gclid=CjwKCAiAzba9BhBhEiwA7glbapXv5khvlN4eHZAhijBUPqukhMp9rSkNL2Rg2ZoDMHny_klc60ZG8RoCY_oQAvD_BwE&utm_campaign=C1000151355N_eg_en_web_searchxxexactandphrasexxbrandpurexx08082022_noon_web_c1000088l_acquisition_sembranded_&utm_medium=cpc&utm_source=C1000088L")
```

```python
iPhones_Elements=fawzi.find_elements("xpath","//div[contains(@class, 'ProductBoxLinkHandler_linkWrapper__b0qZ9')]")
```

```python
iPhones_List=[]
for iPhone in iPhones_Elements:
    iPhone_data={}
    

    iPhone_data["Model"]=iPhone.find_element("xpath",".//h2[contains(@class, 'ProductDetailsSection_title__JorAV')]").get_attribute("title")

    iPhone_data["Current Price"]=iPhone.find_element("xpath",".//strong[contains(@class, 'amount')]").text
    
    try: 
        iPhone_data["Original Price"]=iPhone.find_element("xpath",".//span[contains(@class, 'oldPrice')]").text
    except:
        iPhone_data["Original Price"]=iPhone.find_element("xpath",".//strong[contains(@class, 'amount')]").text
    
    if iPhone_data["Current Price"]==iPhone_data["Original Price"]:
        iPhone_data["Sale"]='-'
    else:
        iPhone_data["Sale"]="On Sale"
        
    iPhone_data["Rating"]=iPhone.find_element("xpath",".//div[contains(@class, 'ProductImageSection_imageSectionFooter__xXCVi')]/div/div/div/div").text
    iPhone_data["Link"]=iPhone.find_element("xpath",".//a").get_attribute("href")   
    
    iPhone_data["Image Link"]=iPhone.find_element("xpath",".//div[contains(@class, 'ProductImage_imageWrapper__C_aHA')]/img").get_attribute("src")
    iPhone_data['Image_Display'] = '<img src="' + iPhone_data["Image Link"] + '" width="100"/>'

    iPhones_List.append(iPhone_data)
```

```python
df = pd.DataFrame.from_dict(iPhones_List) 
```

```python
display(HTML(df.to_html(escape=False)))
```

```python
df.to_csv('iPhone_Prices.csv', index=False) 
```
