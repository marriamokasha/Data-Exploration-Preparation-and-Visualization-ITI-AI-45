```python
from selenium import webdriver
import pandas as pd
```

```python
driver1=webdriver.Chrome()
driver1.get("https://www.amazon.in/gp/bestsellers/books/")
```

```python
Books=driver1.find_elements("xpath","//div[contains(@class, '_cDEzb_iveVideoWrapper_JJ34T')]")
```

```python
Books_List=[]
for book in Books:
    Book_data={}
    Book_data["Rank"]=book.find_element("xpath",".//span[contains(@class, 'zg-bdg-text')]").text
    Book_data["Title"]=book.find_element("xpath",".//a[contains(@class, 'a-link-normal aok-block')]/span").text
    Book_data["Author"]=book.find_element("xpath",".//div[contains(@class, 'a-row a-size-small')]").text
    Book_data["Rating"]=book.find_element("xpath",".//div/a/i/span").get_attribute("textContent")
    Book_data["Review Count"]=book.find_element("xpath",".//div[contains(@class, 'a-icon-row')]").text
    Book_data["Book Type"]=book.find_element("xpath",".//span[contains(@class, 'a-size-small a-color-secondary a-text-normal')]").text
    Book_data["Price"]=book.find_element("xpath",".//span[contains(@class, '_cDEzb_p13n-sc-price_3mJ9Z')]").text
    Book_data["Link"]=book.find_element("xpath",".//a").get_attribute("href")
    Books_List.append(Book_data)
```

```python
df = pd.DataFrame.from_dict(Books_List) 
df
```

```python
df.to_csv('Bestselling Books.csv', index=False) 
```

```python
driver1.quit()
```
